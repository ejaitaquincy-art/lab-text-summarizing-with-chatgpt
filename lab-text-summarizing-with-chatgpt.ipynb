# Text summarizing with ChaptGPT
In this lesson, you will summarize text with a focus on specific topics.

## Setup

````python
from openai import OpenAI
import os

from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv()) # read local .env file

OPENAI_API_KEY  = os.getenv('OPENAI_API_KEY')
````

````python
client = OpenAI(
    # This is the default and can be omitted
    api_key=OPENAI_API_KEY,
)


def get_completion(prompt, model="gpt-3.5-turbo"): # Andrew mentioned that the prompt/ completion paradigm is preferable for this class
    messages = [{"role": "user", "content": prompt}]
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=0, # this is the degree of randomness of the model's output
    )
    return response.choices[0].message.content
````

## Text to summarize

````python
prod_review = """
Got this panda plush toy for my daughter's birthday, \
who loves it and takes it everywhere. It's soft and \ 
super cute, and its face has a friendly look. It's \ 
a bit small for what I paid though. I think there \ 
might be other options that are bigger for the \ 
same price. It arrived a day earlier than expected, \ 
so I got to play with it myself before I gave it \ 
to her.
"""
````

## Summarize with a word/sentence/character limit

````python
prompt = f"""
Your task is to generate a short summary of a product \
review from an ecommerce site. 

Summarize the review below, delimited by triple 
backticks, in at most 30 words. 

Review: ```{prod_review}```
"""

response = get_completion(prompt)
print(response)
````

````python
for i in range(len(reviews)):
    prompt = f"""
    Your task is to generate a short summary of a product \ 
    review from an ecommerce site. 

    Summarize the review below, delimited by triple \
    backticks in at most 20 words. 

    Review: ```{reviews[i]}```
    """

    response = get_completion(prompt)
    print(i, response, "\n")
````

## Report:Prompt Engineering Findings

In this lab, we explored how different prompt formulations influence the quality, focus, and usefulness of generated summaries.

We experimented with multiple prompt styles beyond basic summarization, including emotional tone detection, problem extraction, and internal business rewriting.

The emotional-tone prompt produced insight into customer satisfaction rather than product features. This demonstrated how instructions can shift model attention from facts to sentiment.

The problem-extraction prompt significantly reduced verbosity and removed positive bias. However, it occasionally omitted subtle complaints when they were framed indirectly. This shows that restrictive prompts improve precision but may reduce recall.

The internal-feedback rewrite produced structured and actionable outputs. This version was especially useful in transforming informal reviews into professional business language.

Across experiments, we observed that:

- More specific role framing improves relevance.
- Constraints (word limits, focus areas) reduce hallucinations.
- Extraction prompts are more deterministic than summarization prompts.
- Rewriting prompts improve usability for real-world applications.

Some prompt variations failed when instructions were too vague. In those cases, the model reverted to generic summaries instead of task-specific outputs.

Key learning:

Prompt design directly shapes model reasoning pathways. Clear objectives and constraints lead to more reliable and operational outputs.

Overall, prompt engineering is not about asking better questions, but about defining the task environment precisely.
